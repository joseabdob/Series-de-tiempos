# introduccion
Una serie de tiempo es un conjunto de datos ordenados a lo largo del tiempo, el objetivo de una serie de tiempo es analizar patrones para pronosticar el futuro. El enfoque realizado en el pronostico fue un ARIMA (Autoregressive Integrated Moving Average). Para este pronostico se necesitaron los datos historicos. El diferenciar una serie es eliminar las tendencias para volver la una serie estacionaria lo cual facilita la prediccion. Un principal problema con estos pronosticos son los patrones aleatoria ya que aunque se toman en cuenta los pasados los futuros no se pueden pronosticar.

# Inflacion en Mexico
La inflacion es un fenomeno que se observa en la economia de un pais, es el aumento desordenado de los precios de la mayor parte de los bienes y servicios que se comercian en sus mercados, por un periodo de tiempo prolongado. En México la inflación se mide mediante el Índice Nacional de Precios al Consumidor (INPC). El INPC es un indicador que mide el cambio promedio de precios de un conjunto de bienes y servicios que consume la población. La inflacion afecta a las empresas ya que les cambia su planeacion, los precios cambiantes afectan el precio del producto, la demanda y sus margenes.

```python
# importamos las librerias de siempre
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
inflacion_mex['fecha']= pd.to_datetime(inflacion_mex['fecha'], format= "%Y/%m")
```

```python

inflacion_mex=inflacion_mex.set_index('fecha')
```

```python
#grafica tipo de cambio
inflacion_mex['inflacion_12'].plot()
plt.title('inflacion')
plt.show()
```

```python
#grafica de serie
inflacion_mex.plot()
plt.title('inflacion mexico')
plt.xlabel('fecha')
plt.ylabel('inflacion')
plt.show()
```

```python
inflacion_mex.dropna(inplace=True)
```

```python
result_adf = adfuller(inflacion_mex['inflacion_12'])
print('p-value: %.2f' % result_adf[1])
```

```python
inflacion_mex['inflacion diff'] = inflacion_mex['inflacion_12'].diff()
```

```python
inflacion_mex['inflacion diff'].plot()
plt.title ('inflacion diferenciada')
plt.show()
```

```python
inflacion_mex.index
```

```python
inflacion_mex.index.freq = 'MS'
```

```python
inflacion_mex['inflacion diff'].dropna(inplace=True)
```

```python
result_adf = adfuller(inflacion_mex['inflacion diff'])
print('p-value: %.2f' % result_adf[1])
```

aceptamos

```python
inflacion_mex['inflacion diff'].plot()
```

```python
ic = {}
for ar in range(5):
    for ma in range(5):
        print(f"AR: {ar}, MA: {ma}")
        modelo = ARIMA(inflacion_mex['inflacion diff'].dropna(), order=(ar, 0, ma))
        resultado = modelo.fit()
        ic[(ar, ma)] = [resultado.aic, resultado.bic]

ic = pd.DataFrame(ic, index=["AIC", "BIC"]).T
ic.index = ic.index.set_names(["AR", "MA"])
```

```python
ic.sort_values('AIC', ascending = True)
```

```python
modelo2 = ARIMA(inflacion_mex['inflacion_12'], order=(3, 1, 4)) ##AR (4,4)
resultado2 = modelo2.fit() #calcula
print(resultado2.summary() ) #muestra
```

```python
inflacion_mex['inflacion pronosticado'] = resultado2.predict()
```
